## 모델 분류
### 딥러닝 모델
- MLP, CNN, YOLO
  - 여러 데이터를 학습할 수 있음.
  - 순서에 따라 의미가 달라지는 데이터 학습 가능
  - 결과가 안나옴
  - 순서를 상관하지 않고 학습
- RNN, LSTM
  - 여러 데이터를 학습할 수 있음
  - 순서에 따라 의미가 달라지는 데이터는 학습 가능
  - 결과 잘 나옴
  - 순서를 기억한 채 학습

```python
# 모델링
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
```

```python
model = Sequential()
# input_shape =입력 데이터(문제)의 크기(갯수는 상관하지 않는다. 데이터중 1개를 의미)
# activation = 'tanh' : 임팩트 있는 값들은 기억하지만 아닌것들은 잊는다
model.add(SimpleRNN(units = 30, input_shape = (4,26), activation = 'tanh'))
model.add(Dense(units=26, activation='softmax'))
```

```python
# 학습 방법 설정
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
# 학습
model.fit(X,y,epochs=300)
```
